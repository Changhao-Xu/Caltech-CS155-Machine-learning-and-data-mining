## Problem 4C: Implementation of SGD

Fill in the loss, gradient, and SGD functions according to the guidelines given in the problem statement in order to perform SGD.

```python
def loss(X, Y, w):
    '''
    Calculate the squared loss function.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w: A (D, ) shaped numpy array containing the weight vector.
    
    Outputs:
        The loss evaluated with respect to X, Y, and w.
    '''
    
    #==============================================
    # TODO: Implement the SGD loss function.
    #==============================================
    l = 0
    for i in range(len(X)):
        squared_difference = (Y[i] - np.dot(w, X[i])) ** 2
        l = l + squared_difference
    return l
    pass

def gradient(x, y, w):
    '''
    Calculate the gradient of the loss function with respect to the weight vector w,
    evaluated at a single point (x, y) and weight vector w.
    
    Inputs:
        x: A (D, ) shaped numpy array containing a single data point.
        y: The float label for the data point.
        w: A (D, ) shaped numpy array containing the weight vector.
        
    Output:
        The gradient of the loss with respect to w. 
    '''

    #==============================================
    # TODO: Implement the gradient of the
    # loss function.
    #==============================================    
    g = -2 * (y - (np.dot(w, x))) * x
    return g
    pass

def SGD(X, Y, w_start, eta, N_epochs):
    '''
    Perform SGD using dataset (X, Y), initial weight vector w_start,
    learning rate eta, and N_epochs epochs.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w_start:  A (D, ) shaped numpy array containing the weight vector initialization.
        eta: The step size.
        N_epochs: The number of epochs (iterations) to run SGD.
        
    Outputs:
        W: A (N_epochs, D) shaped array containing the weight vectors from all iterations.
        losses: A (N_epochs, ) shaped array containing the losses from all iterations.
    '''
    
    #==============================================
    # TODO: Implement the SGD algorithm.
    #==============================================    
    W = []
    losses = []
    current_weight = w_start
    for i in range(N_epochs):
        np.random.permutation(X)
        losses.append(loss(X, Y, current_weight))
        for j in range(len(X)):
            current_weight = current_weight - (eta * gradient(X[j], Y[j], current_weight))
        W.append(current_weight)
    return np.asarray(W), losses
    pass
```

### Plotting SGD Convergence

Let's visualize the difference in convergence rates a different way. Plot the loss with respect to epoch (iteration) number for each value of eta on the same graph.

```python
'''Plotting SGD convergence'''

#==============================================
# TODO: For the given learning rates, plot the 
# loss for each epoch.
#==============================================
'''Plotting SGD convergence'''

'''Note that this is the same as the sample solution but I understand it'''

eta_vals = [1e-6, 5e-6, 1e-5, 3e-5, 1e-4]
w_start = [0.01, 0.01]
N_epochs = 1000

legend_handles = []
series_names = []

plt.close('all')
plt.figure()

for eta in eta_vals:
    _, losses = SGD(X1, Y1, w_start, eta, N_epochs)
    name = 'eta: ' + str(eta)
    data_series, = plt.plot(losses, label=name)
    legend_handles.append(data_series)
    series_names.append(name)

plt.legend(legend_handles, series_names, loc = "upper right", bbox_to_anchor=(1, 1))
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('SGD Squared Loss VS Epoch')
plt.show()
```
